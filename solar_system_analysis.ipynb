```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("solar_system_planets.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["planet_type"] = ["Terrestrial", "Terrestrial", "Terrestrial", "Terrestrial",
                     "Gas Giant", "Gas Giant", "Ice Giant", "Ice Giant"]
df
```

```python
plt.figure()
sns.scatterplot(data=df, x="distance_from_sun_au", y="orbital_period_days")
plt.title("Distance from Sun vs Orbital Period")
plt.show()
```

```python
plt.figure()
sns.barplot(data=df, x="planet", y="radius_km")
plt.title("Planet Radius Comparison")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df, x="planet_type", y="gravity_m_s2")
plt.title("Surface Gravity by Planet Type")
plt.show()
```

```python
df.groupby("planet_type")["gravity_m_s2"].mean()
```

## Key Insights

- Orbital period increases significantly with distance from the Sun.
- Gas giants have much larger radii than terrestrial planets.
- Surface gravity varies widely and does not scale linearly with size.
- Simple categorisation supports clearer scientific comparisons.
